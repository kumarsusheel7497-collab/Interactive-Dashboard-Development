```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from google.colab import files
```

```python
df = pd.read_csv("web_traffic.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

for col in df.select_dtypes(include=np.number).columns:
    df[col] = df[col].fillna(df[col].median())

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna("Unknown")
```

```python
print("Total rows:", len(df))

if 'sales' in df.columns:
    print("Total Sales:", df['sales'].sum())

if 'profit' in df.columns:
    print("Total Profit:", df['profit'].sum())

if 'churn' in df.columns:
    print("Total Churn:", df['churn'].astype(str).str.lower().isin(['yes', '1', 'true']).sum())
```

```python
category_col = None

for col in ['category', 'product', 'page', 'region', 'city']:
    if col in df.columns:
        category_col = col
        break

if category_col:
    top_counts = df[category_col].value_counts().head(10).reset_index()
    top_counts.columns = [category_col, 'count']

    fig_bar = px.bar(top_counts, x=category_col, y='count',
                     title=f"Top 10 {category_col.title()} Values")
    fig_bar.show()
```

```python
date_col = None
value_col = None

for col in df.columns:
    if 'date' in col:
        date_col = col
        break

for col in ['sales', 'profit', 'monthly_charges', 'session_duration']:
    if col in df.columns:
        value_col = col
        break

if date_col and value_col:
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
    trend = df.groupby(df[date_col].dt.date)[value_col].sum().reset_index()

    fig_line = px.line(trend, x=date_col, y=value_col,
                       title=f"{value_col.replace('_',' ').title()} Trend Over Time")
    fig_line.show()
```

```python
pie_col = None
for col in ['traffic_source', 'payment_method', 'contract', 'region', 'category']:
    if col in df.columns:
        pie_col = col
        break

if pie_col:
    pie_data = df[pie_col].value_counts().reset_index()
    pie_data.columns = [pie_col, 'count']

    fig_pie = px.pie(pie_data, names=pie_col, values='count',
                     title=f"{pie_col.replace('_',' ').title()} Distribution")
    fig_pie.show()
```

```python
num_col = None
for col in ['sales', 'profit', 'monthly_charges', 'session_duration']:
    if col in df.columns:
        num_col = col
        break

if num_col:
    fig_hist = px.histogram(df, x=num_col, nbins=30,
                            title=f"{num_col.replace('_',' ').title()} Distribution")
    fig_hist.show()
```

```python
html_content = """
<html>
<head>
    <title>Interactive Dashboard</title>
</head>
<body>
    <h1>Interactive Dashboard Development</h1>
    <p>This project presents key insights using interactive charts and summary metrics.</p>
</body>
</html>
"""

with open("interactive_dashboard.html", "w", encoding="utf-8") as f:
    f.write(html_content)

print("interactive_dashboard.html file created successfully.")
```
